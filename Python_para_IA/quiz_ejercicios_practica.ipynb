## ⭐ BLOQUE 6: Ejercicio integrador (nivel quiz real)

**Ejercicio Final** — Dado el dataset Auto-MPG:

1. Elimina las filas con NaN en cualquier columna numérica.
2. Crea una columna `decada` que clasifique el año en: `70s` (años 70-79) o `80s` (años 80-82).
3. Calcula el `mpg` promedio por `decada` y `origen`.
4. Encuentra los 5 coches más eficientes (mayor mpg) de la década de los 80.
5. Usa NumPy para calcular la correlación manual entre `peso` y `mpg`  
   (fórmula: `corr = cov(x,y) / (std(x) * std(y))`).

````python
# Tu código aquí — tómate tu tiempo, este es del nivel del quiz real
````

<details><summary>💡 Ver solución completa</summary>

```python
# 1. Eliminar NaN en columnas numéricas
df_clean = df.dropna(subset=['mpg','cilindros','desplazamiento','potencia','peso','aceleracion'])
print(f"Filas restantes: {len(df_clean)}")

# 2. Columna 'decada'
df_clean = df_clean.copy()
df_clean['decada'] = df_clean['año'].apply(lambda x: '70s' if x < 80 else '80s')

# 3. MPG promedio por decada y origen
display(df_clean.groupby(['decada','origen'])['mpg'].mean().unstack())

# 4. Top 5 coches eficientes de los 80s
top5_80s = df_clean[df_clean['decada']=='80s'].sort_values('mpg', ascending=False).head(5)
display(top5_80s[['nombre','año','mpg','origen']])

# 5. Correlación manual NumPy
peso = df_clean['peso'].values
mpg_v = df_clean['mpg'].values
cov = np.cov(peso, mpg_v)[0,1]
corr_manual = cov / (peso.std() * mpg_v.std())
print(f"Correlación peso-mpg (manual): {corr_manual:.4f}")
print(f"Correlación peso-mpg (pandas): {df_clean['peso'].corr(df_clean['mpg']):.4f}")
```
</details>
