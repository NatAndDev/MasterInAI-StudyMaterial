## Bloque 0 - Recursion (la base de todo)

Antes de entender las tecnicas, necesitas dominar la **recursion**.

### Que es?
Una funcion es **recursiva** si se llama a si misma.
Todo problema recursivo tiene dos partes obligatorias:

```
caso base     -> cuando PARAR (sin esto, bucle infinito)
paso recursivo -> como REDUCIR el problema hacia el caso base
```

### Analogia
Imagina que estas buscando tus llaves en una mochila con bolsillos.
- Si el bolsillo esta vacio -> para (caso base)
- Si tiene mas bolsillos -> busca dentro de cada uno (paso recursivo)


````python
# Ejemplo mas simple: factorial
# 5! = 5 x 4 x 3 x 2 x 1

def factorial(n):
    # CASO BASE: cuando parar
    if n == 0:
        return 1
    # PASO RECURSIVO: reducir hacia el caso base
    return n * factorial(n - 1)

# Traza mental de factorial(4):
#   factorial(4) = 4 * factorial(3)
#                = 4 * 3 * factorial(2)
#                = 4 * 3 * 2 * factorial(1)
#                = 4 * 3 * 2 * 1 * factorial(0)
#                = 4 * 3 * 2 * 1 * 1  <- caso base
#                = 24

for i in range(7):
    print(f"  {i}! = {factorial(i)}")
````

### El arbol de llamadas
Cada llamada recursiva crea un **nodo** en un arbol.
La profundidad del arbol = cuantas veces se llama a si misma.

```
factorial(4)
 |-- factorial(3)
      |-- factorial(2)
           |-- factorial(1)
                |-- factorial(0)  <- caso base, empieza a volver
```

Tu Fibonacci de la AG1 tiene este arbol de llamadas:

```
        Fib(4)
       /      \
    Fib(3)   Fib(2)
    /    \   /    \
 Fib(2) Fib(1) Fib(1) Fib(0)
 /    \
Fib(1) Fib(0)
```

**Problema:** Fib(2), Fib(1), Fib(0) se calculan VARIAS veces. Esto lo resolveremos con DP en el Bloque 4.


````python
# Tu Fibonacci de la AG1 - analizado

def Fibonacci(N: int):
    if N < 2:          # <- CASO BASE: F(0)=1, F(1)=1
        return 1
    else:
        return Fibonacci(N-1) + Fibonacci(N-2)   # <- PASO RECURSIVO

print("Fibonacci(N) para N = 0 a 9:")
for i in range(10):
    print(f"  F({i}) = {Fibonacci(i)}")
````

## Bloque 1 - Divide y Venceras

### La idea central
> "Si el problema es grande, partelo en dos mitades,
>  resuelve cada mitad por separado,
>  y **combina** los resultados."

### Los 3 pasos siempre son los mismos
1. **DIVIDIR** - partir el problema en subproblemas mas pequenos
2. **CONQUISTAR** - resolver cada subproblema (recursivamente)
3. **COMBINAR** - unir las soluciones parciales

### Cuando usar D&C?
- El problema se puede partir en partes del **mismo tipo**
- Las partes son **independientes** (no se solapan)
- Combinar las partes es mas facil que resolver el problema entero


### Torres de Hanoi - de tu AG1

**El problema:** Mover N discos del poste 1 al poste 3, usando el 2 como auxiliar.
**Regla:** nunca poner un disco grande sobre uno pequeno.

**La solucion D&C:**
1. Mover los N-1 discos superiores al poste auxiliar
2. Mover el disco grande al poste destino
3. Mover los N-1 discos del auxiliar al destino

El truco de tu codigo: `auxiliar = 6 - desde - hasta`
Si los postes son 1, 2, 3, entonces 1+2+3 = 6.
Asi el auxiliar siempre es el que queda: si desde=1 y hasta=3, auxiliar = 6-1-3 = 2.


````python
# Tu Torres de Hanoi - con comentarios extendidos

def Torres_Hanoi(N, desde, hasta):
    auxiliar = 6 - desde - hasta   # <- el tercer poste siempre es 6 - los otros dos

    if N == 1:
        # CASO BASE: un solo disco, moverlo directamente
        print(f"  Mover disco {N}: poste {desde} -> poste {hasta}")
    else:
        # PASO 1 (DIVIDIR + CONQUISTAR): mover N-1 discos al auxiliar
        Torres_Hanoi(N-1, desde, auxiliar)

        # PASO 2 (COMBINAR): mover el disco grande al destino
        print(f"  Mover disco {N}: poste {desde} -> poste {hasta}")

        # PASO 3 (CONQUISTAR): mover N-1 discos del auxiliar al destino
        Torres_Hanoi(N-1, auxiliar, hasta)


print("=== Hanoi con 3 discos ===")
Torres_Hanoi(3, 1, 3)
````

````python
# Otro ejemplo D&C clasico: Merge Sort
# Dividir la lista por la mitad, ordenar cada mitad, fusionar

def merge_sort(lista):
    if len(lista) <= 1:
        return lista          # CASO BASE

    mid  = len(lista) // 2
    izq  = merge_sort(lista[:mid])   # CONQUISTAR izquierda
    der  = merge_sort(lista[mid:])   # CONQUISTAR derecha
    return fusionar(izq, der)        # COMBINAR

def fusionar(izq, der):
    resultado = []
    i = j = 0
    while i < len(izq) and j < len(der):
        if izq[i] <= der[j]:
            resultado.append(izq[i]); i += 1
        else:
            resultado.append(der[j]); j += 1
    return resultado + izq[i:] + der[j:]

import random
datos = [random.randint(1, 99) for _ in range(12)]
print(f"Original:  {datos}")
print(f"Ordenada:  {merge_sort(datos)}")
````

## Bloque 2 - Algoritmo Voraz (Greedy)

### La idea central
> "En cada paso, elige la opcion que parece **mejor ahora mismo**,
>  sin mirar hacia atras ni reconsiderar."

### Cuando funciona y cuando falla?
| Situacion | Funciona greedy? |
|-----------|-----------------|
| Monedas canonicas (1,5,10,25) | Si |
| Monedas arbitrarias (1,3,4) | No siempre |
| Mochila fraccional | Si |
| Mochila 0/1 (enteros) | No |

### Los 4 componentes de un algoritmo voraz
1. **Conjunto candidatos** - las opciones disponibles
2. **Funcion de seleccion** - como elegir el "mejor" candidato
3. **Funcion de factibilidad** - la eleccion sigue siendo valida?
4. **Funcion objetivo** - que estamos maximizando/minimizando


````python
# Tu cambio_monedas de la AG1 - explicado

def cambio_monedas(N, SM):
    # N  = cantidad a devolver
    # SM = lista de monedas disponibles (de mayor a menor)
    SOLUCION       = [0] * len(SM)   # cuantas monedas de cada tipo usamos
    ValorAcumulado = 0

    for i, valor in enumerate(SM):
        # FUNCION DE SELECCION: cuantas monedas de este valor caben?
        monedas        = (N - ValorAcumulado) // valor
        SOLUCION[i]    = monedas
        ValorAcumulado += monedas * valor

        if ValorAcumulado == N:
            return SOLUCION     # llegamos al total exacto

    return SOLUCION

# Ejemplo de tu AG1
resultado = cambio_monedas(15, [25, 10, 5, 1])
monedas   = [25, 10, 5, 1]
print("Cambio para 15 centimos:")
for m, c in zip(monedas, resultado):
    if c > 0:
        print(f"  {c} moneda(s) de {m}c  =  {c*m}c")
print(f"  Total: {sum(m*c for m,c in zip(monedas, resultado))}c")
````

````python
# Cuando FALLA el algoritmo voraz?
# Monedas {1, 3, 4} para devolver 6

def cambio_voraz(N, SM):
    usadas, restante = [], N
    for m in sorted(SM, reverse=True):
        while restante >= m:
            usadas.append(m)
            restante -= m
    return usadas

def cambio_optimo(N, SM):
    from itertools import product
    mejor = None
    for r in range(1, N+1):
        for combo in product(SM, repeat=r):
            if sum(combo) == N:
                if mejor is None or len(combo) < len(mejor):
                    mejor = list(combo)
    return mejor or []

SM = [1, 3, 4]
N  = 6

v = cambio_voraz(N, SM)
o = cambio_optimo(N, SM)

print(f"Monedas disponibles: {SM}")
print(f"Cantidad a cambiar:  {N}")
print()
print(f"Greedy:  {sorted(v, reverse=True)}  -> {len(v)} monedas  (4+1+1)")
print(f"Optimo:  {sorted(o, reverse=True)}  -> {len(o)} monedas  (3+3)")
print()
print("Conclusion: greedy NO da el optimo con monedas arbitrarias.")
print("            Para eso necesitamos Programacion Dinamica (Bloque 4).")
````

## Bloque 3 - Backtracking (Vuelta Atras)

### La idea central
> "Construye la solucion paso a paso.
>  Si en algun punto **no puede continuar**, deshaz el ultimo paso
>  y prueba la siguiente opcion."

### La estructura siempre es la misma

```python
def backtrack(solucion, etapa):
    for opcion in opciones_posibles:
        solucion[etapa] = opcion

        if es_prometedora(solucion, etapa):   # vale la pena seguir?
            if etapa == ultima_etapa:
                guardar(solucion)             # solucion completa
            else:
                backtrack(solucion, etapa+1)  # seguir construyendo

        solucion[etapa] = 0                   # DESHACER (backtrack)
```

### Las dos restricciones en N-Reinas
- **Explicita:** no dos reinas en la misma columna -> `SOLUCION.count(valor) <= 1`
- **Implicita:** no dos reinas en la misma diagonal -> `abs(i-j) == abs(Si-Sj)`


````python
# Tu codigo de N-Reinas - explicado linea a linea

def es_prometedora(SOLUCION, etapa):
    # Comprueba si la colocacion hasta 'etapa' es valida
    # SOLUCION[i] = columna donde esta la reina de la fila i
    for i in range(etapa + 1):
        # RESTRICCION EXPLICITA: dos reinas en la misma columna?
        if SOLUCION.count(SOLUCION[i]) > 1:
            return False

        # RESTRICCION IMPLICITA: dos reinas en la misma diagonal?
        for j in range(i + 1, etapa + 1):
            # abs(dif. filas) == abs(dif. columnas)  ->  estan en diagonal
            if abs(i - j) == abs(SOLUCION[i] - SOLUCION[j]):
                return False

    return True


def reinas(N, solucion=None, etapa=0):
    if solucion is None:
        solucion = [0] * N

    for i in range(1, N + 1):
        solucion[etapa] = i              # PROBAR esta columna

        if es_prometedora(solucion, etapa):
            if etapa == N - 1:
                yield solucion[:]        # solucion completa encontrada
            else:
                yield from reinas(N, solucion, etapa + 1)  # profundizar

        solucion[etapa] = 0              # DESHACER (backtrack)


# Mostrar soluciones para N=4
print("Soluciones para 4-Reinas (cada numero = columna en esa fila):\n")
for sol in reinas(4):
    print(" ", sol)
    for fila in range(4):
        print("  " + "".join(" Q " if sol[fila]==col+1 else " . " for col in range(4)))
    print()
````

````python
# Cuantos nodos explora backtracking vs fuerza bruta?

def contar_reinas(N):
    soluciones = [0]
    nodos      = [0]

    def bt(sol, etapa):
        for i in range(1, N + 1):
            nodos[0] += 1
            sol[etapa] = i
            if es_prometedora(sol, etapa):
                if etapa == N - 1:
                    soluciones[0] += 1
                else:
                    bt(sol, etapa + 1)
            sol[etapa] = 0

    bt([0]*N, 0)
    return soluciones[0], nodos[0]

print(f"{'N':>3} | {'Soluciones':>11} | {'Nodos BT':>10} | {'N^N (fuerza bruta)':>20} | {'Ahorro %':>10}")
print("-" * 65)
for n in range(1, 10):
    s, nd = contar_reinas(n)
    fb = n**n
    ahorro = (1 - nd/fb) * 100 if fb > 0 else 0
    print(f"{n:>3} | {s:>11} | {nd:>10} | {fb:>20} | {ahorro:>9.1f}%")
````

## Bloque 4 - Programacion Dinamica (DP)

### La idea central
> "Si la recursion recalcula los mismos subproblemas una y otra vez,
>  **guarda los resultados** en una tabla y reutilizalos."

### DP vs las otras tecnicas
| Tecnica | Subproblemas | Se reusan? | Estrategia |
|---------|-------------|------------|------------|
| D&C | Independientes | No | Partir y combinar |
| Backtracking | Todos posibles | No | Explorar y podar |
| **DP** | **Solapados** | **Si** | **Tabla + recurrencia** |

### Las 4 preguntas para resolver un problema con DP
1. **Cual es el subproblema?** -> definir `dp[i]` o `dp[i][j]`
2. **Cual es la recurrencia?** -> `dp[i] = f(dp[i-1], dp[i-2], ...)`
3. **Cual es el caso base?** -> los valores iniciales de la tabla
4. **Como reconstruir la solucion?** -> tabla RUTA o trazado hacia atras


````python
# Fibonacci: recursion ingenua vs DP
import time

def fib_recursivo(n):
    if n < 2: return 1
    return fib_recursivo(n-1) + fib_recursivo(n-2)

def fib_dp(n):
    # DP bottom-up: construir la tabla desde 0 hasta n
    if n < 2: return 1
    tabla = [0] * (n + 1)
    tabla[0] = 1   # caso base
    tabla[1] = 1   # caso base
    for i in range(2, n + 1):
        tabla[i] = tabla[i-1] + tabla[i-2]   # recurrencia
    return tabla[n]

print(f"{'N':>4} | {'Recursivo (ms)':>16} | {'DP (ms)':>10} | {'Resultado':>12}")
print("-" * 50)
for n in [10, 15, 20, 25, 30]:
    t0 = time.perf_counter(); r1 = fib_recursivo(n); t1 = time.perf_counter()
    t2 = time.perf_counter(); r2 = fib_dp(n);        t3 = time.perf_counter()
    print(f"{n:>4} | {(t1-t0)*1000:>16.3f} | {(t3-t2)*1000:>10.4f} | {r2:>12}")

print("\nDP es ordenes de magnitud mas rapido porque NO recalcula.")
````

### El Viaje por el Rio - de tu AG1

**El problema:** Ir del nodo 0 al nodo 6 pagando el minimo.
No todas las rutas directas existen (tarifa = 999 significa "no existe").

**La recurrencia DP:**
```
PRECIOS[i][j] = min(
    TARIFAS[i][j],                       # ir directo
    PRECIOS[i][k] + TARIFAS[k][j]       # pasar por el nodo k
)  para todo k entre i y j
```

Esto es exactamente el bucle `for k in range(i, j)` de tu codigo.


````python
# Tu codigo del Viaje por el Rio - con comentarios detallados

TARIFAS = [
    [0,   5,   4,   3, 999, 999, 999],
    [999, 0, 999,   2,   3, 999,  11],
    [999, 999,  0,  1, 999,   4,  10],
    [999, 999, 999, 0,   5,   6,   9],
    [999, 999, 999, 999, 0, 999,   4],
    [999, 999, 999, 999, 999, 0,   3],
    [999, 999, 999, 999, 999, 999,  0]
]

def Precios(TARIFAS):
    N      = len(TARIFAS[0])
    INF    = 9999

    # TABLA DP: PRECIOS[i][j] = coste minimo de i a j
    PRECIOS = [[INF]*N for _ in range(N)]
    # TABLA RUTA: RUTA[i][j] = nodo intermedio para ir de i a j
    RUTA    = [[""]*N for _ in range(N)]

    for i in range(N - 1):
        PRECIOS[i][i] = 0     # caso base: ir de i a i cuesta 0
        RUTA[i][i]    = i

        for j in range(i + 1, N):
            PRECIOS[i][j] = TARIFAS[i][j]   # empezar con el coste directo
            RUTA[i][j]    = i

            # RECURRENCIA: hay algun nodo k intermedio mas barato?
            for k in range(i, j):
                coste_via_k = PRECIOS[i][k] + TARIFAS[k][j]
                if coste_via_k < PRECIOS[i][j]:
                    PRECIOS[i][j] = coste_via_k
                    RUTA[i][j]    = k   # registrar que pasamos por k

    return PRECIOS, RUTA

PRECIOS, RUTA = Precios(TARIFAS)

print("Tabla PRECIOS[i][j] = coste minimo de nodo i al nodo j:")
print("     " + "  ".join(f"->N{j}" for j in range(7)))
for i in range(7):
    fila = [f"{PRECIOS[i][j]:>4}" if PRECIOS[i][j] < 9999 else "   x" for j in range(7)]
    print(f"  N{i}: " + "  ".join(fila))

print(f"\nCoste minimo de 0 a 6: {PRECIOS[0][6]}")
print(f"(La tarifa directa era {TARIFAS[0][6]} = 'no existe')")
````

````python
# Reconstruir la ruta optima

def reconstruir_ruta(RUTA, desde, hasta):
    if desde == hasta:
        return [desde]
    intermedio = RUTA[desde][hasta]
    if intermedio == desde:
        return [desde, hasta]
    return reconstruir_ruta(RUTA, desde, intermedio) +            reconstruir_ruta(RUTA, intermedio, hasta)[1:]

ruta = reconstruir_ruta(RUTA, 0, 6)
coste_check = sum(TARIFAS[ruta[i]][ruta[i+1]] for i in range(len(ruta)-1))

print(f"Ruta optima: {' -> '.join(map(str, ruta))}")
print("\nDesglose por tramos:")
for i in range(len(ruta)-1):
    print(f"  Nodo {ruta[i]} -> Nodo {ruta[i+1]}: {TARIFAS[ruta[i]][ruta[i+1]]}")
print(f"  Total: {coste_check}")
````

````python
# Mochila 0/1 con DP - el ejemplo clasico
# (Recuerda: greedy NO funciona aqui)
#
# dp[i][w] = valor maximo usando los primeros i items con capacidad w
#
# Recurrencia:
#   Si el item i NO cabe:  dp[i][w] = dp[i-1][w]
#   Si el item i SI cabe:  dp[i][w] = max(dp[i-1][w],
#                                         dp[i-1][w-peso[i]] + valor[i])

def mochila_01(pesos, valores, capacidad):
    n  = len(pesos)
    dp = [[0]*(capacidad+1) for _ in range(n+1)]

    for i in range(1, n+1):
        for w in range(capacidad+1):
            dp[i][w] = dp[i-1][w]                          # no incluir item i
            if pesos[i-1] <= w:
                dp[i][w] = max(dp[i][w],
                               dp[i-1][w-pesos[i-1]] + valores[i-1])  # incluir

    # Reconstruir que items se eligieron
    elegidos, w = [], capacidad
    for i in range(n, 0, -1):
        if dp[i][w] != dp[i-1][w]:
            elegidos.append(i-1)
            w -= pesos[i-1]

    return dp[n][capacidad], elegidos

pesos   = [2, 3, 4, 5]
valores = [3, 4, 5, 6]
cap     = 8
valor_max, elegidos = mochila_01(pesos, valores, cap)

print(f"Items disponibles:")
for i, (p, v) in enumerate(zip(pesos, valores)):
    marca = " <- ELEGIDO" if i in elegidos else ""
    print(f"  Item {i}: peso={p}, valor={v}{marca}")
print(f"\nCapacidad mochila: {cap}")
print(f"Valor maximo:      {valor_max}")
print(f"Peso total:        {sum(pesos[i] for i in elegidos)}")
````

````python
# EJERCICIO 4
# Implementa la distancia de edicion (Levenshtein) con DP.
# Cuantas operaciones minimas (insertar, borrar, reemplazar)
# se necesitan para convertir s1 en s2?
#
# Recurrencia:
#   dp[0][j] = j  (insertar j caracteres)
#   dp[i][0] = i  (borrar i caracteres)
#   dp[i][j] = dp[i-1][j-1]                     si s1[i-1]==s2[j-1]
#   dp[i][j] = 1 + min(dp[i-1][j],              # borrar
#                       dp[i][j-1],              # insertar
#                       dp[i-1][j-1])            # reemplazar

def distancia_edicion(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0]*(n+1) for _ in range(m+1)]

    # Casos base
    for i in range(m+1): dp[i][0] = i
    for j in range(n+1): dp[0][j] = j

    # TU CODIGO AQUI: rellenar dp[i][j] para i>0, j>0

    return dp[m][n]

# print(distancia_edicion("kitten", "sitting"))   # debe dar 3
# print(distancia_edicion("python", "python"))    # debe dar 0
# print(distancia_edicion("abc", ""))             # debe dar 3
````

## Resumen Final - Cuando usar cada tecnica?

### El arbol de decision

```
El problema se divide en partes INDEPENDIENTES del mismo tipo?
|-- Si -> Los subproblemas se solapan (se recalculan)?
|         |-- Si  -> PROGRAMACION DINAMICA
|         |-- No  -> DIVIDE Y VENCERAS
|-- No -> Existe criterio "mejor ahora" que funciona globalmente?
          |-- Si  -> ALGORITMO VORAZ
          |-- No  -> BACKTRACKING (probar todas las opciones)
```

### Tabla comparativa

| Tecnica | Complejidad tipica | Garantia optimo | Ejemplo en tu AG1 |
|---------|-------------------|-----------------|-------------------|
| D&C | O(n log n) | Si | Torres de Hanoi |
| Greedy | O(n log n) | A veces | Cambio de monedas |
| Backtracking | O(b^d) exponencial | Si (busca todo) | N-Reinas |
| DP | O(n^2) o O(n*W) | Si | Viaje por el rio |
