### Ejercicio 1. Frecuencia de pares
Dada una lista M de números enteros positivos, devolver un diccionario con la frecuencia
de aparición de los números pares de M.


```python
def dpar(M):
    frequencies = {}
    for number in M:
        if number % 2  == 0:
           frequencies[number] = frequencies.get(number, 0) + 1
    return frequencies
```

Prueba

```python
M = [12, 19, 19, 18, 18, 16, 18, 13, 19, 18, 12, 18, 11, 20, 14, 14, 20, 20, 20, 16, 13, 15, 19, 14, 12]
b = dpar(M)
print(b == {12: 3, 18: 5, 16: 2, 20: 4, 14: 3})  # True
```

### Ejercicio 3. Temperaturas nevada
Dados dos diccionarios — personas:ciudad y ciudad:temperatura — devolver la lista
ordenada alfabéticamente de personas que viven en ciudades con temperatura bajo 0ºC.


```python
def PersMayTemp(dPersC, dCiudT):
    result = [person for person, city in dPersC.items() if dCiudT.get(city, 0) < 0]
    return sorted(result)
```

Prueba

```python
dCi = {
    'Manchester': 1.1, 'Madrid': -8.9, 'Gava': 4,
    'Pobla de Segur': -5.6, 'Lleida': -3.2, 'Elche': 2.1,
    'Burgos': -6.0, 'Sant Boi': 4.5
}
dPe = {
    'Pepe': 'Manchester', 'Lionel': 'Gava', 'Mike': 'Sant Boi',
    'Puyol': 'Pobla de Segur', 'Jaime': 'Elche', 'Sergi': 'Lleida',
    'Ernesto': 'Madrid', 'Carlos': 'Burgos'
}
print(PersMayTemp(dPe, dCi))  # ['Carlos', 'Ernesto', 'Puyol', 'Sergi']
```

### Ejercicio 4. Hipertensión
Dado un diccionario persona:[edad, sistólica, diastólica] y una edad límite,
devolver lista alfabética de personas menores de esa edad con hipertensión
(sistólica >= 140 o diastólica >= 90).


```python
def lst_hiper(dicc, edad):
    hipertensePeople = [person for (person, (age, sis, dias)) in dicc.items() if age < edad and (sis >= 140 or dias >= 90)]
    return sorted(hipertensePeople)
```

Prueba

```python
dpers = {
    'Maria':  [40, 135, 90],
    'Nuria':  [63, 141, 92],
    'Jose':   [47, 110, 59],
    'Luis':   [49, 146, 94],
    'Oriol':  [52, 130, 89],
    'Carlos': [65, 125, 89],
    'Pepe':   [70, 130, 92]
}
print(lst_hiper(dpers, 45))  # ['Maria']
print(lst_hiper(dpers, 70))  # ['Luis', 'Maria', 'Nuria']
```

### Ejercicio 5. Nivel de potasio en sangre
Dado un diccionario paciente:[K+] y una lista de 4 valores límite,
devolver un diccionario con la clasificación clínica de cada paciente.


```python
def nivelKsang(dK, lst):
    valor1, valor2, valor3, valor4 = lst
    def classify(potassium):
        if potassium < valor1:
            return 'hipokalemia crítica'
        elif potassium < valor2:
            return 'hipokalemia leve'
        elif potassium <= valor3:
            return 'normal'
        elif potassium <= valor4:
            return 'hiperkalemia moderada'
        else:
            return 'hiperkalemia severa'

    return {patient:classify(potassium) for patient, potassium in dK.items()}
```

Prueba

```python
dK1 = {'Luis': 2.2, 'Carlos': 7.0, 'Laia': 4.0, 'Mikel': 5.5, 'Jordi': 5.2, 'Anna': 3.6, 'Joe': 7.2}
ls1 = [2.0, 3.5, 5.2, 7.0]
dic = nivelKsang(dK1, ls1)
print(dic == {
    'Luis': 'hipokalemia leve',
    'Carlos': 'hiperkalemia moderada',
    'Laia': 'normal',
    'Mikel': 'hiperkalemia moderada',
    'Jordi': 'normal',
    'Anna': 'normal',
    'Joe': 'hiperkalemia severa'
})  # True
```

### Ejercicio 8. Base de datos presión arterial en DataFrame
Dado un diccionario persona:[edad, sistólica, diastólica], devolver un DataFrame con
columnas Nombre, Edad, Sistólica, Diastólica y Diagnóstico (baja/normal/alta).


```python
import pandas as pd

def bloodPressureDiagnosis(systolic, diastolic):
    isLow  = systolic < 90 or diastolic < 60
    isHigh = systolic >= 140 or diastolic >= 90
    if isLow:
        return 'baja'
    elif isHigh:
        return 'alta'
    else:
        return 'normal'

def dfHipertension(dhiper):
    rows = []
    for name, data in dhiper.items():
        age, systolic, diastolic = data
        diagnosis = bloodPressureDiagnosis(systolic, diastolic)
        rows.append([name, age, systolic, diastolic, diagnosis])
    
    columnNames = ['Nombre', 'Edad', 'Sistólica', 'Diastólica', 'Diagnóstico']
    return pd.DataFrame(rows, columns=columnNames)
```

Prueba

```python
dhiper = {
    'Maria':  [40, 135, 90],
    'Nuria':  [63, 141, 92],
    'Jose':   [47, 110, 59],
    'Luis':   [49, 146, 94],
    'Oriol':  [52, 130, 89],
    'Carlos': [65, 125, 89],
    'Pepe':   [70, 130, 92]
}
print(dfHipertension(dhiper))
```
