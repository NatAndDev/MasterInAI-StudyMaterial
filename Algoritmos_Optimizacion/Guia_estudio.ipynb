## 1. ¿Qué es un Algoritmo?

Un **algoritmo** es una secuencia finita, ordenada y bien definida de pasos para resolver un problema o realizar una tarea.

### Propiedades de un buen algoritmo
| Propiedad | Descripción |
|-----------|-------------|
| **Corrección** | Produce el resultado esperado |
| **Eficiencia** | Usa pocos recursos (tiempo y memoria) |
| **Finitud** | Termina en un número finito de pasos |
| **Claridad** | Sus pasos son comprensibles e inequívocos |

### Ejemplo cotidiano
Imagina una **receta de cocina**: ingredientes (entrada) → pasos ordenados → plato terminado (salida). ¡Eso es un algoritmo!


````python
# Ejemplo sencillo: algoritmo para calcular el factorial de n
def factorial(n):
    """Calcula n! = n × (n-1) × ... × 1"""
    if n == 0 or n == 1:   # caso base
        return 1
    resultado = 1
    for i in range(2, n + 1):
        resultado *= i
    return resultado

# Prueba
for n in range(8):
    print(f"  {n}! = {factorial(n)}")
````

## 2. Algoritmos Deterministas vs Probabilísticos

| Tipo | Descripción | Ejemplo |
|------|-------------|---------|
| **Determinista** | Misma entrada → siempre misma salida | Búsqueda binaria |
| **Probabilístico** | Usa aleatoriedad; puede dar distinta salida en cada ejecución | Monte Carlo, Quicksort aleatorio |

### ¿Cuándo usar cada uno?
- **Determinista**: cuando necesitas resultados exactos y reproducibles.
- **Probabilístico**: cuando el espacio de búsqueda es enorme y una buena solución aproximada es suficiente.


````python
import random

# --- Determinista: verificar si un número es primo ---
def es_primo_deterministico(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# --- Probabilístico: test de Miller-Rabin (simplificado) ---
def es_primo_miller_rabin(n, repeticiones=5):
    """Test probabilístico: puede (raramente) dar falsos positivos."""
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for _ in range(repeticiones):
        a = random.randint(2, n - 1)
        if pow(a, n - 1, n) != 1:   # Fermat test simplificado
            return False
    return True  # "probablemente primo"

numeros = [7, 15, 17, 100, 101, 997]
print(f"{'n':>5} | {'Determinista':>14} | {'Miller-Rabin':>14}")
print("-" * 40)
for n in numeros:
    d = es_primo_deterministico(n)
    p = es_primo_miller_rabin(n)
    print(f"{n:>5} | {str(d):>14} | {str(p):>14}")
````

## 3. Algoritmos Exactos vs Heurísticos

| Tipo | Garantía | Velocidad | Uso típico |
|------|----------|-----------|------------|
| **Exacto** | Solución óptima garantizada | Puede ser muy lento | Problemas pequeños |
| **Heurístico** | Buena solución (no necesariamente óptima) | Rápido | Problemas grandes/complejos |

### Analogía
- **Exacto**: explorar *todos* los caminos de un laberinto para encontrar el más corto.
- **Heurístico**: seguir siempre el camino que parece ir más hacia la salida (puede no ser el óptimo, pero es rápido).


````python
# Problema: hallar el subconjunto de números que más se acerca a un objetivo
# --- Exacto: fuerza bruta ---
from itertools import combinations

def subconjunto_exacto(numeros, objetivo):
    mejor = None
    mejor_diff = float('inf')
    for r in range(1, len(numeros) + 1):
        for combo in combinations(numeros, r):
            s = sum(combo)
            if abs(s - objetivo) < mejor_diff:
                mejor_diff = abs(s - objetivo)
                mejor = combo
    return mejor, sum(mejor)

# --- Heurístico: voraz (greedy) ---
def subconjunto_heuristico(numeros, objetivo):
    numeros_ord = sorted(numeros, reverse=True)
    seleccion, total = [], 0
    for n in numeros_ord:
        if total + n <= objetivo:
            seleccion.append(n)
            total += n
    return seleccion, total

nums = [3, 7, 12, 5, 9, 4, 11, 2]
obj  = 20

sol_exacta, suma_e  = subconjunto_exacto(nums, obj)
sol_heur,   suma_h  = subconjunto_heuristico(nums, obj)

print(f"Objetivo : {obj}")
print(f"Exacto   : {sol_exacta}  → suma = {suma_e}")
print(f"Heurístico: {sol_heur} → suma = {suma_h}")
````

## 4. Análisis de Complejidad y Notación Big-O

La **notación Big-O** describe cómo crece el tiempo (o espacio) de un algoritmo cuando el tamaño de entrada `n` aumenta.

### Clases comunes (de mejor a peor)
| Notación | Nombre | Ejemplo |
|----------|--------|---------|
| O(1) | Constante | Acceder a un elemento de lista |
| O(log n) | Logarítmica | Búsqueda binaria |
| O(n) | Lineal | Recorrer una lista |
| O(n log n) | Linealítmica | Merge sort |
| O(n²) | Cuadrática | Bubble sort |
| O(2ⁿ) | Exponencial | Subconjuntos |
| O(n!) | Factorial | Permutaciones |

> ⚠️ Big-O describe el **peor caso** (cota superior asintótica).


````python
import time, random
import matplotlib.pyplot as plt

def o_constante(lst):        return lst[0]
def o_lineal(lst):           return sum(lst)
def o_cuadratica(lst):
    total = 0
    for i in lst:
        for j in lst:
            total += i * j
    return total

tamanios = [100, 500, 1000, 2000, 3000]
tiempos  = {"O(1)": [], "O(n)": [], "O(n²)": []}

for n in tamanios:
    datos = [random.random() for _ in range(n)]
    for nombre, fn in [("O(1)", o_constante), ("O(n)", o_lineal), ("O(n²)", o_cuadratica)]:
        t0 = time.perf_counter()
        fn(datos)
        tiempos[nombre].append(time.perf_counter() - t0)

plt.figure(figsize=(9, 4))
for nombre, vals in tiempos.items():
    plt.plot(tamanios, vals, marker='o', label=nombre)
plt.xlabel("Tamaño de entrada n")
plt.ylabel("Tiempo (segundos)")
plt.title("Crecimiento del tiempo según complejidad")
plt.legend(); plt.tight_layout(); plt.show()
print("✓ Observa cómo O(n²) crece mucho más rápido.")
````

## 5. Clases de Complejidad: P, NP, NP-Completo, NP-Duro

```
          ┌─────────────────────────────┐
          │           NP-Duro           │
          │   ┌─────────────────────┐   │
          │   │    NP-Completo      │   │
          │   │   ┌─────────────┐   │   │
          │   │   │     NP      │   │   │
          │   │   │  ┌───────┐  │   │   │
          │   │   │  │   P   │  │   │   │
          │   │   │  └───────┘  │   │   │
          │   │   └─────────────┘   │   │
          │   └─────────────────────┘   │
          └─────────────────────────────┘
```

| Clase | Definición informal | Ejemplo |
|-------|---------------------|---------|
| **P** | Solucionable en tiempo polinomial | Ordenar, búsqueda binaria |
| **NP** | Verificable en tiempo polinomial | Problema del viajante (verificar una ruta) |
| **NP-Completo** | NP y tan difícil como cualquier problema NP | SAT, Knapsack 0/1 |
| **NP-Duro** | Al menos tan difícil como NP-Completo (puede no ser NP) | Halting Problem |

> 🔑 La pregunta abierta más famosa en computación: **¿P = NP?** (Premio Clay de $1 millón sin resolver.)


````python
# Demostración conceptual: verificar una solución de Knapsack es fácil (NP)
# pero encontrar la óptima por fuerza bruta es exponencial

def valor_mochila(items, seleccion, capacidad):
    """Verifica y calcula el valor de una selección de ítems."""
    peso_total  = sum(items[i][1] for i in range(len(items)) if seleccion[i])
    valor_total = sum(items[i][0] for i in range(len(items)) if seleccion[i])
    if peso_total > capacidad:
        return -1, peso_total  # no válida
    return valor_total, peso_total

# items = (valor, peso)
items = [(60, 10), (100, 20), (120, 30)]
cap   = 50

# Una solección posible: tomar ítem 1 y 2
seleccion = [0, 1, 1]
v, p = valor_mochila(items, seleccion, cap)
print(f"Selección {seleccion}: valor={v}, peso={p}  → {'válida ✓' if v>=0 else 'inválida ✗'}")

# Verificación es O(n) — fácil (NP)
# Fuerza bruta para encontrar óptimo es O(2^n) — difícil
print(f"
Con n={len(items)} ítems → {2**len(items)} posibles subconjuntos que evaluar por fuerza bruta.")
````

## 6. Algoritmos de Ordenamiento

| Algoritmo | Mejor caso | Caso promedio | Peor caso | Estable |
|-----------|-----------|---------------|-----------|---------|
| Insertion Sort | O(n) | O(n²) | O(n²) | ✅ |
| Bubble Sort | O(n) | O(n²) | O(n²) | ✅ |
| Selection Sort | O(n²) | O(n²) | O(n²) | ❌ |
| Merge Sort | O(n log n) | O(n log n) | O(n log n) | ✅ |
| Heap Sort | O(n log n) | O(n log n) | O(n log n) | ❌ |
| Quick Sort | O(n log n) | O(n log n) | O(n²) | ❌ |
| Radix Sort | O(nk) | O(nk) | O(nk) | ✅ |


````python
# ── Insertion Sort ──────────────────────────────────────────
def insertion_sort(arr):
    a = arr[:]
    for i in range(1, len(a)):
        clave = a[i]
        j = i - 1
        while j >= 0 and a[j] > clave:
            a[j+1] = a[j]; j -= 1
        a[j+1] = clave
    return a

# ── Bubble Sort ─────────────────────────────────────────────
def bubble_sort(arr):
    a = arr[:]
    n = len(a)
    for i in range(n):
        intercambiado = False
        for j in range(n - i - 1):
            if a[j] > a[j+1]:
                a[j], a[j+1] = a[j+1], a[j]; intercambiado = True
        if not intercambiado: break
    return a

# ── Selection Sort ──────────────────────────────────────────
def selection_sort(arr):
    a = arr[:]
    for i in range(len(a)):
        min_idx = i
        for j in range(i+1, len(a)):
            if a[j] < a[min_idx]: min_idx = j
        a[i], a[min_idx] = a[min_idx], a[i]
    return a

# ── Merge Sort ──────────────────────────────────────────────
def merge_sort(arr):
    if len(arr) <= 1: return arr[:]
    mid = len(arr) // 2
    L, R = merge_sort(arr[:mid]), merge_sort(arr[mid:])
    resultado, i, j = [], 0, 0
    while i < len(L) and j < len(R):
        if L[i] <= R[j]: resultado.append(L[i]); i += 1
        else:            resultado.append(R[j]); j += 1
    return resultado + L[i:] + R[j:]

# ── Heap Sort ───────────────────────────────────────────────
def heap_sort(arr):
    import heapq
    h = arr[:]
    heapq.heapify(h)
    return [heapq.heappop(h) for _ in range(len(h))]

# ── Quick Sort ──────────────────────────────────────────────
def quick_sort(arr):
    if len(arr) <= 1: return arr[:]
    pivot = arr[len(arr) // 2]
    izq   = [x for x in arr if x < pivot]
    medio = [x for x in arr if x == pivot]
    der   = [x for x in arr if x > pivot]
    return quick_sort(izq) + medio + quick_sort(der)

# ── Radix Sort ──────────────────────────────────────────────
def radix_sort(arr):
    if not arr: return arr
    maximo = max(arr); exp = 1
    a = arr[:]
    while maximo // exp > 0:
        cubetas = [[] for _ in range(10)]
        for n in a: cubetas[(n // exp) % 10].append(n)
        a = [x for c in cubetas for x in c]
        exp *= 10
    return a

# ── Demostración ────────────────────────────────────────────
import random
datos = [random.randint(1, 99) for _ in range(12)]
print(f"Original : {datos}")
print(f"Insertion: {insertion_sort(datos)}")
print(f"Bubble   : {bubble_sort(datos)}")
print(f"Selection: {selection_sort(datos)}")
print(f"Merge    : {merge_sort(datos)}")
print(f"Heap     : {heap_sort(datos)}")
print(f"Quick    : {quick_sort(datos)}")
print(f"Radix    : {radix_sort(datos)}")
````

````python
# Comparación de tiempos de ordenamiento
import time, random
import matplotlib.pyplot as plt

algoritmos = {
    "Insertion": insertion_sort,
    "Bubble":    bubble_sort,
    "Selection": selection_sort,
    "Merge":     merge_sort,
    "Heap":      heap_sort,
    "Quick":     quick_sort,
    "Radix":     radix_sort,
}

tamanios = [100, 500, 1000, 2000]
resultados = {k: [] for k in algoritmos}

for n in tamanios:
    datos = [random.randint(0, 9999) for _ in range(n)]
    for nombre, fn in algoritmos.items():
        t0 = time.perf_counter()
        fn(datos)
        resultados[nombre].append(time.perf_counter() - t0)

plt.figure(figsize=(10, 5))
for nombre, vals in resultados.items():
    plt.plot(tamanios, vals, marker='o', label=nombre)
plt.xlabel("Tamaño n"); plt.ylabel("Tiempo (s)")
plt.title("Comparación de algoritmos de ordenamiento")
plt.legend(bbox_to_anchor=(1.05, 1)); plt.tight_layout(); plt.show()
````

## 7. Técnicas de Diseño de Algoritmos

### 7.1 Algoritmo Voraz (Greedy)
Toma la **mejor decisión local** en cada paso, esperando llegar a un óptimo global.
- ✅ Rápido y simple
- ⚠️ No siempre da el óptimo global


````python
# Problema del cambio de monedas (Greedy)
def cambio_voraz(monto, monedas):
    monedas = sorted(monedas, reverse=True)
    seleccion = []
    for m in monedas:
        while monto >= m:
            seleccion.append(m); monto -= m
    return seleccion

monedas_euro = [200, 100, 50, 20, 10, 5, 2, 1]  # céntimos
monto = 387  # 3,87 €
resultado = cambio_voraz(monto, monedas_euro)
print(f"Cambio para {monto} céntimos: {resultado}")
print(f"Total de monedas usadas: {len(resultado)}")
````

### 7.2 Backtracking (Vuelta Atrás)
Explora soluciones de forma sistemática, **deshaciendo** las decisiones cuando llega a un callejón sin salida.


````python
# N-Reinas: colocar N reinas en un tablero N×N sin que se ataquen
def n_reinas(n):
    soluciones = []
    def backtrack(tablero, fila):
        if fila == n:
            soluciones.append(tablero[:])
            return
        for col in range(n):
            if es_seguro(tablero, fila, col, n):
                tablero.append(col)
                backtrack(tablero, fila + 1)
                tablero.pop()   # ← BACKTRACK: deshacer

    def es_seguro(tablero, fila, col, n):
        for r, c in enumerate(tablero):
            if c == col or abs(c - col) == abs(r - fila):
                return False
        return True

    backtrack([], 0)
    return soluciones

for n in [4, 5, 6, 7, 8]:
    sols = n_reinas(n)
    print(f"  {n}-Reinas: {len(sols)} soluciones")

# Visualizar una solución para n=5
sol = n_reinas(5)[0]
print(f"
Una solución para 5-Reinas: {sol}")
for fila, col in enumerate(sol):
    print("  " + "· " * col + "♛ " + "· " * (5 - col - 1))
````

### 7.3 Divide y Vencerás
Divide el problema en **subproblemas más pequeños**, resuelve cada uno, y **combina** las soluciones.


````python
# Búsqueda binaria — Divide y Vencerás
def busqueda_binaria(arr, objetivo, izq=0, der=None):
    if der is None: der = len(arr) - 1
    if izq > der:   return -1            # no encontrado

    mid = (izq + der) // 2
    if arr[mid] == objetivo:   return mid
    elif arr[mid] < objetivo:  return busqueda_binaria(arr, objetivo, mid+1, der)
    else:                      return busqueda_binaria(arr, objetivo, izq, mid-1)

lista = list(range(0, 100, 3))   # [0, 3, 6, ..., 99]
for objetivo in [33, 50, 99, 100]:
    idx = busqueda_binaria(lista, objetivo)
    if idx >= 0:
        print(f"  Encontré {objetivo} en índice {idx} ✓")
    else:
        print(f"  {objetivo} no está en la lista ✗")
````

### 7.4 Programación Dinámica (DP)
Resuelve subproblemas **una sola vez** y almacena sus resultados (memoización) para reutilizarlos.


````python
# Fibonacci con y sin programación dinámica
import sys, time

# Sin DP (recursión ingenua) — exponencial O(2^n)
def fib_recursivo(n):
    if n <= 1: return n
    return fib_recursivo(n-1) + fib_recursivo(n-2)

# Con DP (memoización) — lineal O(n)
def fib_dp(n, memo={}):
    if n in memo:   return memo[n]
    if n <= 1:      return n
    memo[n] = fib_dp(n-1, memo) + fib_dp(n-2, memo)
    return memo[n]

# Comparar tiempos
print("n  | sin DP  | con DP")
print("---|---------|--------")
for n in [20, 25, 30, 35]:
    t0 = time.perf_counter(); r1 = fib_recursivo(n); t1 = time.perf_counter()
    t2 = time.perf_counter(); r2 = fib_dp(n, {});    t3 = time.perf_counter()
    print(f"{n:2} | {(t1-t0)*1000:6.2f} ms | {(t3-t2)*1000:.4f} ms  (resultado: {r2})")

# ── Knapsack 0/1 con DP ──────────────────────────────────────
def knapsack_dp(pesos, valores, capacidad):
    n = len(pesos)
    dp = [[0]*(capacidad+1) for _ in range(n+1)]
    for i in range(1, n+1):
        for w in range(capacidad+1):
            dp[i][w] = dp[i-1][w]
            if pesos[i-1] <= w:
                dp[i][w] = max(dp[i][w], dp[i-1][w-pesos[i-1]] + valores[i-1])
    return dp[n][capacidad]

pesos   = [2, 3, 4, 5]
valores = [3, 4, 5, 6]
cap     = 8
print(f"
Knapsack DP — valor máximo: {knapsack_dp(pesos, valores, cap)}")
````

## 8. Problemas Clásicos de Optimización

### 8.1 Problema de la Mochila (Knapsack)
Dado un conjunto de ítems con peso y valor, ¿cuáles meter en una mochila de capacidad W para **maximizar el valor total**?

### 8.2 Problema del Viajante (TSP)
Un vendedor debe visitar N ciudades **exactamente una vez** y volver al origen, minimizando la distancia total.

### 8.3 Programación Entera (IP)
Optimización donde las variables deben ser **enteras** (no continuas). Se usa en planificación, logística, etc.


````python
# ── Knapsack fraccional (Greedy óptimo) ─────────────────────
def knapsack_fraccional(items, capacidad):
    """items = lista de (valor, peso)"""
    items_ord = sorted(items, key=lambda x: x[0]/x[1], reverse=True)
    valor_total, peso_total = 0, 0
    detalles = []
    for v, p in items_ord:
        if peso_total + p <= capacidad:
            peso_total  += p
            valor_total += v
            detalles.append((v, p, 1.0))
        else:
            fraccion     = (capacidad - peso_total) / p
            valor_total += v * fraccion
            detalles.append((v, p, fraccion))
            break
    return valor_total, detalles

items = [(60, 10), (100, 20), (120, 30), (80, 15)]  # (valor, peso)
valor, detalles = knapsack_fraccional(items, 50)
print("Mochila fraccional:")
print(f"  {'Valor':>6} {'Peso':>5} {'Fracción':>10}")
for v, p, f in detalles:
    print(f"  {v:>6} {p:>5} {f:>10.2%}")
print(f"  Valor total obtenido: {valor:.2f}")
````

````python
# ── TSP: Greedy del vecino más cercano ──────────────────────
import math, random, matplotlib.pyplot as plt

def distancia(a, b):
    return math.sqrt((a[0]-b[0])**2 + (a[1]-b[1])**2)

def tsp_vecino_cercano(ciudades):
    n      = len(ciudades)
    no_vis = set(range(n))
    ruta   = [0];  no_vis.remove(0)
    while no_vis:
        actual    = ruta[-1]
        siguiente = min(no_vis, key=lambda c: distancia(ciudades[actual], ciudades[c]))
        ruta.append(siguiente); no_vis.remove(siguiente)
    ruta.append(ruta[0])  # volver al origen
    return ruta

# Generar ciudades aleatorias
random.seed(42)
n_ciudades = 10
ciudades   = [(random.uniform(0, 100), random.uniform(0, 100)) for _ in range(n_ciudades)]
ruta       = tsp_vecino_cercano(ciudades)
dist_total = sum(distancia(ciudades[ruta[i]], ciudades[ruta[i+1]]) for i in range(len(ruta)-1))

# Visualizar
x = [c[0] for c in ciudades]; y = [c[1] for c in ciudades]
plt.figure(figsize=(6, 5))
plt.scatter(x, y, s=100, zorder=5, color='steelblue')
for i, (cx, cy) in enumerate(ciudades):
    plt.annotate(f" {i}", (cx, cy), fontsize=9)
for i in range(len(ruta)-1):
    a, b = ciudades[ruta[i]], ciudades[ruta[i+1]]
    plt.plot([a[0], b[0]], [a[1], b[1]], 'r-', alpha=0.6)
plt.title(f"TSP — Vecino más cercano
Distancia total: {dist_total:.1f}")
plt.tight_layout(); plt.show()
print(f"Ruta: {' → '.join(map(str, ruta))}")
print(f"Distancia total: {dist_total:.2f}")
````

## 9. Algoritmos de Búsqueda

### BFS vs DFS

| | BFS (anchura) | DFS (profundidad) |
|--|--------------|-------------------|
| **Estructura** | Cola (FIFO) | Pila (LIFO) / recursión |
| **Garantiza camino más corto** | ✅ (grafos no ponderados) | ❌ |
| **Uso de memoria** | Alto (almacena nivel completo) | Bajo |
| **Ideal para** | Caminos más cortos, niveles | Detectar ciclos, topología |

### Branch and Bound
Explora el árbol de búsqueda **podando** ramas que no pueden mejorar la mejor solución encontrada.


````python
from collections import deque

# Grafo de ejemplo (lista de adyacencia)
grafo = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E'],
}

# ── BFS ───────────────────────────────────────────────────
def bfs(grafo, inicio, objetivo):
    cola      = deque([(inicio, [inicio])])
    visitados = set()
    while cola:
        nodo, camino = cola.popleft()
        if nodo == objetivo:
            return camino
        if nodo in visitados: continue
        visitados.add(nodo)
        for vecino in grafo[nodo]:
            cola.append((vecino, camino + [vecino]))
    return None

# ── DFS ───────────────────────────────────────────────────
def dfs(grafo, inicio, objetivo, visitados=None, camino=None):
    if visitados is None: visitados = set()
    if camino    is None: camino    = [inicio]
    visitados.add(inicio)
    if inicio == objetivo: return camino
    for vecino in grafo[inicio]:
        if vecino not in visitados:
            resultado = dfs(grafo, vecino, objetivo, visitados, camino + [vecino])
            if resultado: return resultado
    return None

print("BFS A → F:", ' → '.join(bfs(grafo, 'A', 'F')))
print("DFS A → F:", ' → '.join(dfs(grafo, 'A', 'F')))
````

````python
# ── Branch and Bound: Knapsack 0/1 ─────────────────────────
# Estimación del límite superior: relajación fraccional
def upper_bound(idx, peso_act, valor_act, items, capacidad):
    if peso_act >= capacidad: return 0
    bound = valor_act
    w     = peso_act
    for i in range(idx, len(items)):
        v, p = items[i]
        if w + p <= capacidad:
            w += p; bound += v
        else:
            bound += (capacidad - w) * (v / p)
            break
    return bound

def knapsack_branch_bound(items, capacidad):
    items_ord = sorted(items, key=lambda x: x[0]/x[1], reverse=True)
    mejor     = [0]

    def bb(idx, peso, valor):
        if idx == len(items_ord) or peso == capacidad:
            mejor[0] = max(mejor[0], valor)
            return
        # Rama "incluir"
        v, p = items_ord[idx]
        if peso + p <= capacidad:
            bb(idx+1, peso+p, valor+v)
        # Rama "excluir" — sólo si puede superar la mejor
        if upper_bound(idx+1, peso, valor, items_ord, capacidad) > mejor[0]:
            bb(idx+1, peso, valor)

    bb(0, 0, 0)
    return mejor[0]

items = [(10, 5), (6, 3), (5, 4), (9, 7), (3, 2)]
cap   = 10
print(f"Branch & Bound Knapsack → valor máximo: {knapsack_branch_bound(items, cap)}")
print(f"DP Knapsack             → valor máximo: {knapsack_dp([p for v,p in items],[v for v,p in items],cap)}")
````

## 10. Descenso de Gradiente

El **descenso de gradiente** es un algoritmo de optimización que minimiza una función moviéndose en la **dirección opuesta al gradiente** (la pendiente).

```
x_nuevo = x_actual - α × ∇f(x_actual)
```

- **α (tasa de aprendizaje)**: cuánto nos movemos en cada paso.
  - Muy grande → puede divergir.
  - Muy pequeña → converge lento.


````python
import numpy as np
import matplotlib.pyplot as plt

# Función a minimizar: f(x) = x⁴ - 4x² + x + 3
# Derivada:          f'(x) = 4x³ - 8x + 1

def f(x):   return x**4 - 4*x**2 + x + 3
def df(x):  return 4*x**3 - 8*x + 1

def descenso_gradiente(x0, alpha=0.01, iteraciones=500):
    x = x0
    historia = [x]
    for _ in range(iteraciones):
        x = x - alpha * df(x)
        historia.append(x)
    return x, historia

x0 = 2.5   # punto de inicio
x_min, hist = descenso_gradiente(x0, alpha=0.01, iteraciones=300)

# Visualizar
xs = np.linspace(-2.5, 2.5, 300)
plt.figure(figsize=(9, 4))

plt.subplot(1, 2, 1)
plt.plot(xs, f(xs), 'b-', lw=2)
hist_y = [f(x) for x in hist]
plt.plot(hist[::10], hist_y[::10], 'ro-', ms=4, alpha=0.6, label='Pasos')
plt.scatter([x_min], [f(x_min)], s=100, color='green', zorder=5, label=f'Mínimo≈{x_min:.3f}')
plt.title("Función y trayectoria"); plt.xlabel("x"); plt.ylabel("f(x)"); plt.legend()

plt.subplot(1, 2, 2)
plt.plot(hist_y)
plt.title("Convergencia del valor de f(x)"); plt.xlabel("Iteración"); plt.ylabel("f(x)")

plt.tight_layout(); plt.show()
print(f"Mínimo encontrado en x ≈ {x_min:.4f},  f(x) ≈ {f(x_min):.4f}")
````

## 11. Heurísticas y Metaheurísticas

Las **metaheurísticas** son estrategias de alto nivel que guían la búsqueda de soluciones de calidad en problemas difíciles.

| Metaheurística | Inspiración | Mecanismo clave |
|----------------|-------------|-----------------|
| **Búsqueda Tabú** | — | Lista de movimientos prohibidos |
| **Recocido Simulado** | Enfriamiento de metales | Acepta soluciones peores con probabilidad decreciente |
| **PSO** | Parvadas de aves | Partículas que siguen al mejor global y local |
| **GRASP** | — | Construcción aleatoria + búsqueda local |


````python
# ── Recocido Simulado (TSP simplificado) ────────────────────
import math, random, copy

def dist_ruta(ciudades, ruta):
    return sum(distancia(ciudades[ruta[i]], ciudades[ruta[(i+1) % len(ruta)]]) for i in range(len(ruta)))

def recocido_simulado_tsp(ciudades, T_ini=1000, T_fin=1, alpha=0.995, iters=5000):
    n       = len(ciudades)
    ruta    = list(range(n))
    random.shuffle(ruta)
    mejor   = ruta[:]
    T       = T_ini
    historia = []

    for it in range(iters):
        # Generar vecino: intercambiar dos ciudades
        i, j = random.sample(range(n), 2)
        nueva = ruta[:]
        nueva[i], nueva[j] = nueva[j], nueva[i]

        delta = dist_ruta(ciudades, nueva) - dist_ruta(ciudades, ruta)
        if delta < 0 or random.random() < math.exp(-delta / T):
            ruta = nueva
            if dist_ruta(ciudades, ruta) < dist_ruta(ciudades, mejor):
                mejor = ruta[:]

        T = max(T * alpha, T_fin)
        if it % 200 == 0:
            historia.append(dist_ruta(ciudades, mejor))

    return mejor, dist_ruta(ciudades, mejor), historia

random.seed(7)
ciudades_sa = [(random.uniform(0, 100), random.uniform(0, 100)) for _ in range(15)]
ruta_sa, dist_sa, hist_sa = recocido_simulado_tsp(ciudades_sa)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
for i in range(len(ruta_sa)):
    a = ciudades_sa[ruta_sa[i]]; b = ciudades_sa[ruta_sa[(i+1) % len(ruta_sa)]]
    plt.plot([a[0], b[0]], [a[1], b[1]], 'b-', alpha=0.5)
xs2 = [c[0] for c in ciudades_sa]; ys2 = [c[1] for c in ciudades_sa]
plt.scatter(xs2, ys2, s=60, zorder=5)
plt.title(f"Recocido Simulado
Distancia: {dist_sa:.1f}")

plt.subplot(1, 2, 2)
plt.plot(hist_sa, 'r-')
plt.title("Convergencia"); plt.xlabel("Checkpoint"); plt.ylabel("Mejor distancia")

plt.tight_layout(); plt.show()
````

````python
# ── PSO: Optimización por Enjambre de Partículas ────────────
# Minimizar f(x, y) = x² + y²  (mínimo en origen)
import numpy as np
import matplotlib.pyplot as plt

def funcion_objetivo(pos):
    return pos[0]**2 + pos[1]**2

class Particula:
    def __init__(self, dim=2, lim=10):
        self.pos     = np.random.uniform(-lim, lim, dim)
        self.vel     = np.random.uniform(-1,   1,   dim)
        self.mejor   = self.pos.copy()
        self.f_mejor = funcion_objetivo(self.pos)

def pso(n_part=30, dim=2, iters=100, w=0.5, c1=1.5, c2=1.5):
    particulas    = [Particula(dim) for _ in range(n_part)]
    mejor_global  = min(particulas, key=lambda p: p.f_mejor).mejor.copy()
    historia_f    = []

    for _ in range(iters):
        for p in particulas:
            r1, r2 = np.random.rand(dim), np.random.rand(dim)
            p.vel  = w * p.vel + c1*r1*(p.mejor - p.pos) + c2*r2*(mejor_global - p.pos)
            p.pos += p.vel
            f      = funcion_objetivo(p.pos)
            if f < p.f_mejor:
                p.f_mejor = f; p.mejor = p.pos.copy()
                if f < funcion_objetivo(mejor_global):
                    mejor_global = p.pos.copy()
        historia_f.append(funcion_objetivo(mejor_global))

    return mejor_global, historia_f

np.random.seed(0)
sol_pso, hist_pso = pso()
plt.figure(figsize=(6, 3))
plt.plot(hist_pso, 'purple')
plt.title("PSO — Convergencia (minimizar x²+y²)"); plt.xlabel("Iteración"); plt.ylabel("f(mejor global)")
plt.tight_layout(); plt.show()
print(f"Mejor solución encontrada: x={sol_pso[0]:.4f}, y={sol_pso[1]:.4f}")
print(f"Valor de f: {funcion_objetivo(sol_pso):.6f}  (óptimo = 0)")
````

````python
# ── Búsqueda Tabú (Knapsack) ────────────────────────────────
def busqueda_tabu_knapsack(items, cap, iters=200, tabu_len=10):
    n = len(items)
    # Solución inicial aleatoria válida
    sol = [0] * n
    for i in range(n):
        if sum(items[j][1]*sol[j] for j in range(n)) + items[i][1] <= cap:
            sol[i] = 1

    def valor(s):
        if sum(items[j][1]*s[j] for j in range(n)) > cap: return -1
        return sum(items[j][0]*s[j] for j in range(n))

    mejor     = sol[:]
    mejor_val = valor(mejor)
    tabu      = []
    hist      = [mejor_val]

    for _ in range(iters):
        # Generar vecinos: cambiar un bit
        vecinos = []
        for i in range(n):
            v = sol[:]
            v[i] = 1 - v[i]
            if i not in tabu and valor(v) >= 0:
                vecinos.append((valor(v), i, v))
        if not vecinos: break
        vecinos.sort(reverse=True)
        _, mov, sol = vecinos[0]
        tabu.append(mov)
        if len(tabu) > tabu_len: tabu.pop(0)
        if valor(sol) > mejor_val:
            mejor     = sol[:]
            mejor_val = valor(mejor)
        hist.append(mejor_val)

    return mejor, mejor_val, hist

items_tabu = [(10,5),(6,3),(5,4),(9,7),(3,2),(8,4),(7,6),(4,3)]
cap_tabu   = 15
sol_t, val_t, hist_t = busqueda_tabu_knapsack(items_tabu, cap_tabu)

plt.figure(figsize=(6, 3))
plt.plot(hist_t, 'teal')
plt.title("Búsqueda Tabú — Knapsack"); plt.xlabel("Iteración"); plt.ylabel("Mejor valor")
plt.tight_layout(); plt.show()
print(f"Mejor solución: {sol_t}  (valor = {val_t})")
````

## 12. Algoritmos Evolutivos

Inspiran en la **evolución biológica** para explorar espacios de búsqueda enormes.

### 12.1 Algoritmos Genéticos (GA)
```
Población inicial → Selección → Cruce → Mutación → Nueva generación → (repetir)
```

### 12.2 Optimización por Colonia de Hormigas (ACO)
Inspirado en cómo las hormigas encuentran el camino más corto usando **feromonas**.


````python
# ── Algoritmo Genético: Maximizar una función ───────────────
import random, numpy as np
import matplotlib.pyplot as plt

# Maximizar f(x) = x * sin(10π*x) + 1  en [−1, 2]
def f_obj(x):
    return x * np.sin(10 * np.pi * x) + 1.0

def ga(pop_size=60, generaciones=100, tasa_mut=0.02):
    # Codificación real
    pop  = [random.uniform(-1, 2) for _ in range(pop_size)]
    hist = []

    for gen in range(generaciones):
        fitnesses = [f_obj(x) for x in pop]
        # Selección por torneo
        def torneo():
            a, b = random.choices(range(pop_size), k=2)
            return pop[a] if fitnesses[a] > fitnesses[b] else pop[b]

        nueva = []
        while len(nueva) < pop_size:
            p1, p2 = torneo(), torneo()
            # Cruce aritmético
            alpha = random.random()
            hijo  = alpha * p1 + (1 - alpha) * p2
            # Mutación gaussiana
            if random.random() < tasa_mut:
                hijo += random.gauss(0, 0.1)
            hijo = max(-1, min(2, hijo))  # mantener en rango
            nueva.append(hijo)

        pop  = nueva
        hist.append(max(f_obj(x) for x in pop))

    mejor_x = max(pop, key=f_obj)
    return mejor_x, f_obj(mejor_x), hist

random.seed(42)
mejor_x, mejor_f, hist_ga = ga()

fig, axes = plt.subplots(1, 2, figsize=(11, 4))

xs = np.linspace(-1, 2, 500)
axes[0].plot(xs, f_obj(xs), 'b-', lw=1.5, label='f(x)')
axes[0].scatter([mejor_x], [mejor_f], s=120, color='red', zorder=5,
                label=f'Mejor: x={mejor_x:.3f}, f={mejor_f:.3f}')
axes[0].set_title("Función y mejor solución GA"); axes[0].legend()

axes[1].plot(hist_ga, 'green')
axes[1].set_title("Convergencia del GA"); axes[1].set_xlabel("Generación"); axes[1].set_ylabel("Mejor fitness")

plt.tight_layout(); plt.show()
print(f"Mejor x encontrado: {mejor_x:.4f}")
print(f"f(x) = {mejor_f:.4f}")
````

````python
# ── Colonia de Hormigas (ACO) — TSP simplificado ────────────
import numpy as np, random, matplotlib.pyplot as plt

np.random.seed(1); random.seed(1)
N = 8  # ciudades
coords = np.random.rand(N, 2) * 100

# Matriz de distancias
D = np.array([[np.sqrt(sum((coords[i]-coords[j])**2)) for j in range(N)] for i in range(N)])
np.fill_diagonal(D, np.inf)

def aco_tsp(n_hormigas=20, iters=80, alpha=1, beta=3, rho=0.3, Q=100):
    feromonas = np.ones((N, N))
    visibilidad = 1.0 / D
    mejor_ruta, mejor_dist = None, float('inf')
    hist = []

    for _ in range(iters):
        rutas = []
        for _ in range(n_hormigas):
            ruta = [random.randint(0, N-1)]
            no_vis = set(range(N)) - {ruta[0]}
            while no_vis:
                i = ruta[-1]
                probs = np.array([(feromonas[i][j]**alpha * visibilidad[i][j]**beta)
                                   if j in no_vis else 0 for j in range(N)])
                probs /= probs.sum()
                sig = np.random.choice(N, p=probs)
                ruta.append(sig); no_vis.remove(sig)
            rutas.append(ruta)

        # Actualizar feromonas
        feromonas *= (1 - rho)
        for ruta in rutas:
            d = sum(D[ruta[i]][ruta[(i+1)%N]] for i in range(N))
            for i in range(N):
                feromonas[ruta[i]][ruta[(i+1)%N]] += Q / d
            if d < mejor_dist:
                mejor_dist = d; mejor_ruta = ruta[:]

        hist.append(mejor_dist)

    return mejor_ruta, mejor_dist, hist

ruta_aco, dist_aco, hist_aco = aco_tsp()

fig, axes = plt.subplots(1, 2, figsize=(11, 4))

axes[0].scatter(coords[:,0], coords[:,1], s=80, zorder=5)
for i, (x, y) in enumerate(coords):
    axes[0].annotate(f" {i}", (x, y), fontsize=9)
for i in range(N):
    a, b = coords[ruta_aco[i]], coords[ruta_aco[(i+1)%N]]
    axes[0].plot([a[0],b[0]], [a[1],b[1]], 'r-', alpha=0.7)
axes[0].set_title(f"ACO — TSP ({N} ciudades)
Distancia: {dist_aco:.1f}")

axes[1].plot(hist_aco, 'orange')
axes[1].set_title("Convergencia ACO"); axes[1].set_xlabel("Iteración"); axes[1].set_ylabel("Mejor distancia")

plt.tight_layout(); plt.show()
print(f"Mejor ruta: {' → '.join(map(str, ruta_aco + [ruta_aco[0]]))}  (distancia = {dist_aco:.2f})")
````
